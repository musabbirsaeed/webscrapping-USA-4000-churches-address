```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
import pickle
```

```python
url = "http://www.usachurches.org/"
state_data = requests.get(url, headers={'User-Agent': 'XYZ/3.0'})

state_data
```

```python
state_soup = BeautifulSoup(state_data.content)

#print(state_soup)
```

```python
links= state_soup.select("div.states a")
links_list = [link['href'] for link in links]
links_list
```

```python
church_city_list=[]
for link in links_list:
    cities = requests.get(url+link, headers={'User-Agent': 'XYZ/3.0'})
    soup_city = BeautifulSoup(cities.content)
    cities_links = soup_city.select("div.citylinks a")
    church_city = [link['href'] for link in cities_links][1:]
    church_city_list.append(church_city)
print(church_city_list)
```

```python
flat_list_city = [item for sublist in church_city_list for item in sublist]
```

```python
flat_list_city
```

```python
url_al_ga = flat_list_city[100]
pagenumber = ['','20/', '40/', '60/', '80/']
[url_al_ga + i for i in pagenumber]
```

```python
church_link_list = []
pagenumber = ['','20/', '40/', '60/', '80/']
#pagenumber = ['']
for i in flat_list_city:
    for j in pagenumber:
        test_church = requests.get(i+j, headers={'User-Agent': 'XYZ/3.0'})
        test_church = BeautifulSoup(test_church.content)
        for link in test_church.select("strong a"):
            if link['href'] not in church_link_list:
                church_link_list.append(link['href'])
        
print(church_link_list)
```
